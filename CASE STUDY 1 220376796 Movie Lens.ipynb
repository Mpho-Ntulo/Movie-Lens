# QUESTION 3 - hint:(string manipulation and Heatmap)

```python
uf1 = data3.release_date.str.slice(-4)  #performing string manipulation
uf1.head()
```

```python
data4 = pd.read_csv('u.genre', sep='|')
data4.head()
```

```python
data4.rename(columns={'unknown':'genre',},inplace=True)
UD = pd.merge(uf1,data4['genre'],how='outer',left_index=True,right_index=True)
UD.head()
```

```python
ud = pd.merge(data,data3,left_on='userID',right_on='movie_id')
ud.head()
```

```python
UD = pd.merge(UD,ud['rating'],how='outer',left_index=True,right_index=True)
UD.head()
```

```python
UD.duplicated().sum()
```

```python
UD.drop_duplicates(inplace=True)
```

```python
UD.duplicated().sum()
```

```python
UD.isnull().sum()
```

```python
UD.dropna(inplace=True)
```

```python
UD.head()
```

```python
UD.tail()
```

```python
plt.figure(figsize=(12,6))
UD = UD.pivot( 'genre','release_date','rating')
sns.heatmap(UD, annot=True, linewidths=0.9,center = 1,vmin=0,vmax=6)
plt.show()
```
